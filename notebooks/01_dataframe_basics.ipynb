```python
from pyspark.sql import SparkSession

# Initialize Spark session
spark = SparkSession.builder.appName("OrderJoinSumFilter").getOrCreate()

# More sample Orders
order = spark.createDataFrame([
    (1, "2024-01-01", "C100"),
    (2, "2024-01-02", "C101"),
    (3, "2024-01-03", "C102"),
    (4, "2024-01-04", "C103"),
    (5, "2024-01-05", "C100"),
], ["order_id", "order_date", "customer_id"])

# More sample Order Lines
order_line = spark.createDataFrame([
    (1, 101, 2),   # order_id, product_id, quantity
    (1, 102, 1),
    (2, 101, 5),
    (2, 103, 4),
    (3, 101, 1),
    (3, 104, 3),
    (4, 102, 2),
    (4, 103, 2),
    (4, 105, 5),
    (5, 101, 3),
    (5, 105, 1)
], ["order_id", "product_id", "quantity"])

# Expanded Product list
product = spark.createDataFrame([
    (101, "Milk"),
    (102, "Bread"),
    (103, "Butter"),
    (104, "Cheese"),
    (105, "Eggs")
], ["product_id", "product_name"])
order.show()
order_line.show()
product.show()
```
